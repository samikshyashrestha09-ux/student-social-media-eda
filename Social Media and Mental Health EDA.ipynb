# DATA VISUALIZATION

```python
plt.figure(figsize = (7,5))
plt.title('Age Distribution')
sns.histplot(x= 'Age', data = df)
```

**The majority of students in the dataset are 21 years old, suggesting the dataset is largely representative of undergraduate-level students.**

```python
plt.title('Gender Distribution')
sns.countplot(x='Gender', data = df)
```

**The dataset is nearly balanced in terms of gender, with male students slightly outnumbering female students (~2600 vs ~2350).**

```python
df['Country'].nunique()
```

```python
top_countries = df['Country'].value_counts().head(10)
plt.figure(figsize = (15,5))
plt.title('Top 10 Countries')
plt.bar(top_countries.index, top_countries.values)
plt.show()
```

**The dataset is globally distributed, with the majority of students grouped under 'Other' countries, followed by India and the USA as the most represented individual countries.**

```python
plt.title('Average Daily Usage_Hours Distribution')
sns.histplot(x= 'Avg_Daily_Usage_Hours', data = df)
```

**Social media usage among students is concentrated between 4 to 6 hours per day, with the peak around 5 hours, suggesting a majority of students spend a significant portion of their day on social media.**

```python
plt.title('Daily_Unlocks Distribution')
sns.histplot(x= 'Daily_Unlocks', data = df)
```

**Students unlock their devices between 150 to 200 times per day on average, with the distribution being slightly right-skewed — indicating that while most students fall in this range, a smaller group unlocks their phone as many as 270 times a day, roughly once every 3-4 minutes.**

```python
sns.countplot(x= 'Academic_Level', data = df)
```

**The majority of students in the dataset are undergraduates, followed by graduate and high school students, suggesting the dataset is most representative of university-level social media behavior.**

```python
plt.figure(figsize = (15,5))
plt.title('Most Used Platforms')
sns.countplot(x= 'Most_Used_Platform', data = df)
```

**Instagram is the most used platform among students, followed by TikTok and Facebook. Regional platforms such as KakaoTalk, WeChat, and LINE have significantly lower usage, which aligns with the dataset being predominantly composed of students from Western and South Asian countries.**

```python
plt.title('Average Daily Usage Hours VS Mental Health Score')
sns.scatterplot(x = 'Avg_Daily_Usage_Hours', y = 'Mental_Health_Score', data = df)
```

**There is a clear negative relationship between average daily social media usage and mental health score. As usage hours increase, mental health scores declines.**

```python
plt.title('Average Daily Usage Hours VS Stress Level')
sns.boxplot(data=df, x="Stress_Level", y="Avg_Daily_Usage_Hours", order = ["Low", "Medium", "High", "Very High"])
```

**Students with low stress levels report significantly fewer average daily social media usage hours compared to those with high or very high stress levels.**

```python
plt.title('Mental Health Score VS Stress Level')
custom_order = ["Low", "Medium", "High", "Very High"]
sns.barplot(data=df, x="Stress_Level", y="Mental_Health_Score", order=custom_order)
```

**Students with low stress levels report the highest mental health scores, which gradually decline as stress levels increase. This confirms a strong inverse relationship between stress and mental wellbeing — students experiencing very high stress have noticeably lower mental health scores.**

```python
plt.title('Mental Health Score VS Physical Activity Hours')
sns.scatterplot(x = 'Physical_Activity_Hours', y = 'Mental_Health_Score', data = df)
```

**Students who engage in more physical activity report higher mental health scores, suggesting that regular exercise plays a positive role in student mental wellbeing — even in the context of heavy social media usage.**

```python
plt.title('Stress Level VS Sleep Hours Per Night')
custom_order = ["Low", "Medium", "High", "Very High"]
sns.violinplot(x='Stress_Level', y='Sleep_Hours_Per_Night', data=df,  order=custom_order)
```

**Students with low stress levels sleep between 8 to 9 hours per night, while those with very high stress sleep only around 5 to 6 hours. The narrowing and downward shift of the violin shape as stress increases confirms that poor sleep is strongly associated with higher stress levels among students.**

```python
plt.figure(figsize = (15,5))
plt.title('Most Used Platform VS Mental Health Score')
sns.boxplot(data=df, x="Most_Used_Platform", y="Mental_Health_Score")
```

**Mental health scores are broadly similar across the most widely used platforms, ranging between 6 to 7. Platforms with very few users such as LINE and VKontakte appear to show higher scores, but their small sample sizes make these results unreliable. This suggests that the choice of platform matters less than the amount of time spent on social media overall.**

```python
plt.title(' Regression plot showing Average Daily Usage Hours VS Mental Health Score')
sns.regplot(x = 'Avg_Daily_Usage_Hours', y = 'Mental_Health_Score', data = df)
```

**The regression plot confirms the negative relationship between average daily social media usage and mental health score. The downward sloping trend line indicates that for every additional hour spent on social media, mental health scores tend to decline.**

```python
plt.title('Correlation Between Student Lifestyle and Mental Health Indicators')
numeric_cols = df.select_dtypes(include='number')
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
```

**The correlation heatmap reveals that average daily social media usage has a strong negative correlation with study hours (-0.88), sleep hours (-0.81), and mental health score (-0.82). Conversely, study hours and sleep hours show a positive correlation with mental health score (0.75 and 0.77 respectively). This confirms that heavy social media usage is associated with less studying, poorer sleep, and worse mental wellbeing among students.**

# CONCLUSION
**Heavy social media usage is associated with less studying, less sleep, and worse mental health scores among students.**
